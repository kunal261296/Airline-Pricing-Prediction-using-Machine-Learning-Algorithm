```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
data=pd.read_csv('airline.csv')
print(data.columns)
data = data.drop(["TravelMonth","SeatsPremium","PitchPremium","WidthPremium", "PricePremium", "PriceRelative", "SeatsTotal", "PitchDifference","WidthDifference"], axis=1)
print(data.columns)


from statsmodels.formula.api import ols
y=data['PriceEconomy']
X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=0.4,random_state=0)
pred_model=ols("""PriceEconomy~Airline+Aircraft+FlightDuration+IsInternational+SeatsEconomy+PitchEconomy+WidthEconomy+PercentPremiumSeats""",data=X_train).fit()
summary=pred_model.summary()
print(summary)
```

```python
pred=pred_model.predict(X_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': pred}) 
df
```

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
data=pd.read_csv('airline.csv')
print(data.columns)
data = data.drop(["TravelMonth","SeatsEconomy","PitchEconomy","WidthEconomy", "PriceEconomy", "PriceRelative", "SeatsTotal", "PitchDifference","WidthDifference"], axis=1)
print(data.columns)

from statsmodels.formula.api import ols
y=data['PricePremium']
X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=0.4,random_state=0)
pred_model=ols("""PricePremium~Airline+Aircraft+FlightDuration+IsInternational+SeatsPremium+PitchPremium+WidthPremium+PercentPremiumSeats""",data=X_train).fit()
summary=pred_model.summary()
print(summary)
```

```python
pred=pred_model.predict(X_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': pred}) 
df
```
